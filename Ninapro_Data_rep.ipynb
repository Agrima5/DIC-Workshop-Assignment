```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

df = pd.read_csv("Ninapro_DB1[1].csv")

print(df.head())
```

```python
'''Data Representation'''
```

```python
#Bar Graph
```

```python
plt.figure(figsize=(20,10))
df.exercise.value_counts().plot(kind = 'bar');
plt.title("Exercise")
plt.ylabel("Frequency")
plt.xlabel("Exercise Number")
plt.legend()
plt.show()
```

```python
#Histogram
```

```python
plt.hist(df[df["exercise"]==1]["glove_0"], color="blue",label="Ex. 1", alpha= 0.7, density= True)
plt.hist(df[df["exercise"]==2]["glove_0"], color="red",label="Ex. 2", alpha= 0.7, density= True)
plt.hist(df[df["exercise"]==3]["glove_0"], color="yellow",label="Ex. 3", alpha= 0.7, density= True)
plt.title("Histogarm of Glove Value for the 3 categories of exercises")
plt.legend()
plt.xlabel("Value in glove #0")
plt.ylabel("Frequency")
plt.show()
```

```python
#Pie Chart
```

```python
(df["repetition"].value_counts()).plot.pie(autopct="%.1f%%", rotatelabels=False, radius=1)
plt.title("Pie chart of Repetition values")
plt.show()
```
